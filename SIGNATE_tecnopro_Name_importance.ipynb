```python
df = pd.merge(test,train,how="left",on = "point_name")
```

```python
df_nonExist = df[df["N_10"].isna()]
```

```python
df["chara_Num"] = df_nonExist["chara_Num"] = train["chara_Num"] = 0
df["head_Chara"] = df_nonExist["head_Chara"] = train["head_Chara"] = "initial value"

for i in df.index.values:
  df["chara_Num"][i] = len(df["point_name"][i])
  df["head_Chara"][i] = df["point_name"][i][0]

for i in df_nonExist.index.values:
  df_nonExist["chara_Num"][i] = len(df_nonExist["point_name"][i])
  df_nonExist["head_Chara"][i] = df_nonExist["point_name"][i][0]

for i in train.index.values:
  train["chara_Num"][i] = len(train["point_name"][i])
  train["head_Chara"][i] = train["point_name"][i][0]
```

```python
temp_chara_mean = df.iloc[0:0]

for chara in df["head_Chara"].unique():
  N_10_chara_mean = pd.DataFrame(df.groupby(["head_Chara"]).mean()["N_10"]).query(f"head_Chara=='{chara}'").values[0][0]
  E_10_chara_mean = pd.DataFrame(df.groupby(["head_Chara"]).mean()["E_10"]).query(f"head_Chara=='{chara}'").values[0][0]
  temp_chara_mean_ = df[df["head_Chara"] == chara].fillna({"N_10": N_10_chara_mean, "E_10": E_10_chara_mean})
  temp_chara_mean = pd.concat([temp_chara_mean, temp_chara_mean_])
```
